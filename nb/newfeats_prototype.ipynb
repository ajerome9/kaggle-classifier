```python
def get_zip_stats(df):
    df.loc[:, 'ZIP123INT'] = (df.loc[:, 'VAR_0241']/100).astype('int')
    df.loc[:, 'ZIP9'] = df.VAR_0212.astype(str).str[:9]
    
    s1 = df.groupby('VAR_0241')['target'].mean()
    s2 = df.groupby('VAR_0241')['target'].sum()
    s3 = df.groupby('VAR_0241')['target'].count()
    s3diff = s3-s2
    s1.name='ZIP_RESP_RATE'
    s2.name='ZIP_RESP_COUNT'
    s3.name='ZIP_SENT_COUNT'
    s3diff.name='ZIP_RESP_DIFF'

    s4 = df.groupby('ZIP123INT')['target'].mean()
    s5 = df.groupby('ZIP123INT')['target'].sum()
    s6 = df.groupby('ZIP123INT')['target'].count()
    s6diff = s6-s5
    s4.name='ZIP123_RESP_RATE'
    s5.name='ZIP123_RESP_COUNT'
    s6.name='ZIP123_SENT_COUNT'
    s6diff.name='ZIP123_RESP_DIFF'
    
    s7 = df.groupby('ZIP9')['target'].mean()
    s8 = df.groupby('ZIP9')['target'].sum()
    s9 = df.groupby('ZIP9')['target'].count()
    s9diff = s9-s8
    s7.name='ZIP9_RESP_RATE'
    s8.name='ZIP9_RESP_COUNT'
    s9.name='ZIP9_SENT_COUNT'
    s9diff.name='ZIP9_RESP_DIFF'
    
    zipdf = pd.concat([s1, s2, s3, s3diff], axis=1)
    zip123df = pd.concat([s4, s5, s6, s6diff], axis=1)
    zip9df = pd.concat([s7, s8, s9, s9diff], axis=1)
    zip9df.loc[zip9df.index == '-1.0', ['ZIP9_RESP_RATE', 'ZIP9_RESP_COUNT', 'ZIP9_SENT_COUNT', 'ZIP9_RESP_DIFF']] = -99999
    
    dfnew = pd.merge(zipdf, df, right_on='VAR_0241', left_index=True, how='right')
    dfnew = pd.merge(zip123df, dfnew, right_on='ZIP123INT', left_index=True, how='right')
    dfnew = pd.merge(zip9df, dfnew, right_on='ZIP9', left_index=True, how='right')
    dfnew.set_index('ID', inplace=True)
    dfnew.reset_index(inplace=True)
    
    dfnew = dfnew.drop(['ZIP123INT'], axis=1)
    dfnew = dfnew.drop(['ZIP9'], axis=1)
    
    return dfnew

strain = pd.read_csv(conf.basepath+'train.csv', dtype=pickle.load(open(conf.dtypesf, 'rb')))
strain.fillna(-1, inplace=True)
tnew = get_zip_stats(strain)
```
