```python
from collections import defaultdict
import operator
import numpy as np

with open('/Users/aa/lab/common-repo/kaggle/springleaf/nogit/056-topdel/run1/output-xgbdump.txt', 'r') as f:
    lines = f.readlines()

def get_bw(s, start, end):
    return ((s.split(start))[1].split(end)[0])
    
allgain = defaultdict(float)
count = defaultdict(int)
poslists = defaultdict(list)
for line in lines:
    line = line.strip()
    if not line.startswith('booster'):
        parts = line.split(':')
        pos, line = parts[0], parts[1]
        if line.startswith('['):
            fcomp = get_bw(line, '[', ']')
            feat = get_bw(fcomp, 'f', '<' if '<' in fcomp else '>')
            gain = get_bw(line, 'gain=', ',')
            #print(line, feat, gain)
            allgain[feat] += float(gain)
            count[feat] += 1
            poslists[feat].append(int(pos))

def pos_stats(feat, poslists):
    a = np.array(poslists[feat])
    return np.mean(a), np.std(a)
    
sortedByTotGain = False
if sortedByTotGain:
    totgain = sum(allgain.values())
    allgain = sorted(allgain.items(), key=operator.itemgetter(1), reverse=True)
    print('gain sorted\nfeat, pct, gain, count, gainpercount, posmean, posstd')
    for item in allgain:
        print('f{}, {:.2f}, {:.2f}, {}, {:.2f}, {:.2f}, {:.2f}'.format(item[0], 100*item[1]/totgain, 
                                                                       item[1], count[item[0]], item[1]/count[item[0]], 
                                                                       *pos_stats(item[0], poslists)))
else:
    totcount = sum(count.values())
    count = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
    print('count sorted\nfeat, pct, count, gain, gainpercount, posmean, posstd')
    for item in count:
        print('f{}, {:.2f}, {}, {:.2f}, {:.2f}, {:.2f}, {:.2f}'.format(item[0], 100*item[1]/totcount, 
                                                                       item[1], allgain[item[0]], allgain[item[0]]/item[1],
                                                                       *pos_stats(item[0], poslists)))
```
