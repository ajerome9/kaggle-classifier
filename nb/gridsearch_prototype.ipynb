```python
import itertools
p = (('max_depth', [4,5]),
     ('eta', [0.01, 0.1]),
     ('gamma', [1, 2, 3]))

pvalues = [e[1] for e in p]
pnames = [e[0] for e in p]

for comb in itertools.product(*pvalues):
    print({k:v for k, v in zip(pnames, comb)})
```

```python
numcombs = 1
for values in pvalues:
    numcombs *= len(values)
print(numcombs)
```
